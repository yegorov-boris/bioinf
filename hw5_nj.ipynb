```python
matrix1 = {
    ('A', 'B'): 16,
    ('A', 'C'): 16,
    ('A', 'D'): 10,
    ('B', 'C'): 8,
    ('B', 'D'): 8,
    ('C', 'D'): 4,
}

matrix2 = {
    ('A', 'B'): 5,
    ('A', 'C'): 4,
    ('A', 'D'): 7,
    ('A', 'E'): 6,
    ('A', 'F'): 8,
    ('B', 'C'): 7,
    ('B', 'D'): 10,
    ('B', 'E'): 9,
    ('B', 'F'): 11,
    ('C', 'D'): 7,
    ('C', 'E'): 6,
    ('C', 'F'): 8,
    ('D', 'E'): 5,
    ('D', 'F'): 9,
    ('E', 'F'): 8,
}
```

```python
def nj(matrix, nodes):
    if len(matrix) == 1:
        for (i, j), v in matrix.items():
            node_i = nodes.get(i, i)
            node_j = nodes.get(j, j)
            return f'({node_i}:{round(v, 2)},{node_j}:{round(v, 2)})'
    
    alph = dict({})
    for (i, j), v in matrix.items():
        alph[j] = alph.get(j, 0) + v
        alph[i] = alph.get(i, 0) + v
    
    n = len(alph)
    q = {}
    for k, v in matrix.items():
        i, j = k
        q[k] = v - (alph[i]+alph[j]-2*v)/(n-2)
        
    k_min = min(q, key=q.get)
    v_min = matrix[k_min]
    i_min, j_min = k_min
    j_new = i_min+j_min
    
    new_matrix = {}
    for k, v in matrix.items():
        i, j = k
        if (i not in k_min) and (j not in k_min):
            new_matrix[k] = v
            
    for i, _ in alph.items():
        if i not in k_min:
           new_matrix[(i, j_new)] = 0.5*(
               matrix.get((i, i_min), 0) +
               matrix.get((i_min, i), 0) +
               matrix.get((i, j_min), 0) +
               matrix.get((j_min, i), 0) -
               v_min
           )
    
    d_i = 0.5*(v_min+(alph[i_min]-alph[j_min])/(n-2))
    d_j = 0.5*(v_min+(alph[j_min]-alph[i_min])/(n-2))
    node_i = nodes.get(i_min, i_min)
    node_j = nodes.get(j_min, j_min)
    nodes[j_new] = f'({node_i}:{round(d_i, 2)},{node_j}:{round(d_j, 2)})'
    
#     print(k_min, v_min)
    
#     print('......')
#     for k, v in alph.items():
#         print(k, v)
    
#     print('-----')
#     for k, v in q.items():
#         print(k, v)
    
#     print('=====')
#     for k, v in new_matrix.items():
#         print(k, v)
        
#     print('+++++')
#     for k, v in nodes.items():
#         print(k, v)
    
    return nj(new_matrix, nodes)

test_number = input()
if test_number == '1':
    print(nj(matrix1, {}))
if test_number == '2':
    print(nj(matrix2, {}))
```
